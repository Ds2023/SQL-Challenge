**Tasks**

Write a SQL script that creates the following summary tables (can be tables or views):

a) District_summary.

i) District Name

ii) Region Name

iii) No. of Individual Beneficiaries (1HH = 6 Individuals)

iv) No. of Individual Beneficiaries / Total District Population

```python
%sql select * from jurisdiction_hierarchy limit 5;
```

```python
%%sql SELECT
    jh.jurisdiction AS District_Name,
    jh.parent_jurisdiction AS Region_Name,
    SUM(CASE WHEN bpd.beneficiary_type = 'Individuals' THEN bpd.beneficiaries
             WHEN bpd.beneficiary_type = 'Households' THEN bpd.beneficiaries * 6
             ELSE 0 END) AS No_of_Individual_Beneficiaries,
    CAST(SUM(CASE WHEN bpd.beneficiary_type = 'Individuals' THEN bpd.beneficiaries
                  WHEN bpd.beneficiary_type = 'Households' THEN bpd.beneficiaries * 6
                  ELSE 0 END) AS REAL) / vl.total_population AS Ratio_Individuals_to_Population
FROM
    jurisdiction_hierarchy jh
JOIN
    village_locations vl ON jh.jurisdiction = vl.village
LEFT JOIN
    beneficiary_partner_data bpd ON jh.jurisdiction = bpd.village
WHERE
    jh.jurisdiction_type = 'District'
GROUP BY
    jh.jurisdiction, jh.parent_jurisdiction, vl.total_population;
```

```python
%%sql SELECT
    b.partner_id,
    b.partner,
    COUNT(DISTINCT v.village) AS num_villages_reached,
    COUNT(DISTINCT j.jurisdiction) AS num_districts_reached
FROM
    beneficiary_partner_data b
JOIN
    village_locations v ON b.village = v.village
JOIN
    jurisdiction_hierarchy j ON v.village = j.jurisdiction
GROUP BY
    b.partner_id, b.partner;
```

```python
%%sql SELECT
    v.village_id,
    v.village,
    v.latitude,
    v.longitude,
    v.total_population,
    j.jurisdiction_id,
    j.jurisdiction_type,
    j.parent_jurisdiction,
    b.partner_id,
    b.partner,
    b.beneficiaries,
    b.beneficiary_type
FROM
    village_locations v
JOIN
    jurisdiction_hierarchy j ON v.village = j.jurisdiction
JOIN
    beneficiary_partner_data b ON v.village = b.village;
```
